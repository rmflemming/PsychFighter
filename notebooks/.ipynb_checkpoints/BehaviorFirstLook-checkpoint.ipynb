# Behavioral data preprocessing/ 1st looks  
Here, I'll take a quick look at some of the behavioral data collected so far, just to get an idea of what we're working with! Will also do some rudimentary preprocessing.

## Set up & Imports

```python
# Packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from jupyterthemes import jtplot
jtplot.style()
```

```python
# Data
data_dir = '~/Documents/GitHub/PsychFighter/'
data = 'log_101 .txt'
cols = ['p1_strike','p1_block','ai_strike','ai_block','p1_win']
df = pd.read_csv(data_dir+data,names=cols)
df.head()
```

## Preprocessing  
Notice that some values in each row are 9999. This is not a valid time and is used as a filler where a number is not present. So, we will go head and replace 9999 with None.

```python
df[df == 9999] = None
df.head()
```

## Basic Insights
What could be some interesting information worth visualizing? How often the person attacks vs parries? Who makes the first move? Win rate? Let's start to explore some of these sequentially.

```python
# Simple Stats:
winrate = df.p1_win.mean(); 
p_att = df.p1_strike.count() / len(df.p1_strike);
p_par = df.p1_block.count() / len(df.p1_block);
p_nan = 1 - p_att - p_par;

print('Win rate = ', winrate,'\nP(Strike) = ', p_att, '\nP(Block) = ', p_par, '\nP(No action) = ', p_nan)
```

We see that the participant won a little less than half of the games. They were slightly biased to attack, and only did not make a move (~fast enough~) in 3 trials of the 50 total. These are some handy summary stats, but maybe we are interested more in the timecourse of the participants behavior.

```python
plt.figure(figsize=(9,6));
df.p1_strike.plot();
df.p1_block.plot();
df.ai_strike.plot(color='red');
df.ai_block.plot(color='orange')
df.winshades = df.p1_win * 1.75
df.winshades.plot(kind='bar',width=1,color='purple',alpha=0.2);
plt.xlabel('trials'); plt.xticks(np.linspace(0,49,11),np.linspace(1,50,11,dtype=int));
plt.ylabel('trial time'); plt.ylim([0,1.75]); plt.yticks(np.linspace(0,1.75,8),np.linspace(0,1.75,8));
plt.title('Timecourse of a single game',fontsize=22);
plt.legend(bbox_to_anchor = (1.25,1));
```

This may be a little hard to interpret, but from this figure, you may get the sense that the participant wins more frequently when they attack than when they block. Let's see if this is the case:

```python
# make masks
nullmask = df.isnull()
attkmask = nullmask.p1_strike
parrmask = nullmask.p1_block
```

```python
p_w_att = df.p1_win[~attkmask].mean()
p_w_par = df.p1_win[~parrmask].mean()
print('Attack win rate = ', p_w_att,'\nParry win rate = ', p_w_par)
```
